```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt, dates as mdates
pd.set_option('display.max_rows', None)
```

```python
eth_hunt = pd.read_csv('eth_hunt.csv')
```

```python
# GET THE ALL SELLING ETH TO SWAP FOR HUNT 
eth_hunt_swaps = eth_hunt[eth_hunt['event_name'] == 'Swap']
eth_hunt_swaps = eth_hunt_swaps[eth_hunt_swaps['buy/sell'] == 'Sell']
```

```python
eth_hunt_swaps
```

```python
# FINDING DIFFERENCE TO KNOW THE PNL
pnl = eth_hunt_swaps['HUNT_in($)'] - eth_hunt_swaps['ETH_out($)']
pnl
```

```python
#DISPLAY IN LINEPLOT
graph = sns.lineplot(x = eth_hunt_swaps['block_timestamp'],y = pnl,color = 'red')
graph.set(xlabel ="DATE", ylabel = "PNL($)")
```
